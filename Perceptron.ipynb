```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("placement-dataset.csv")
df.head()
```

```python
df.drop(['city'], axis=1,inplace=True)
```

```python
df.head()
```

```python
df['cgpa'].unique()
df['cgpa']=df['cgpa'].replace(np.nan,5)
df['cgpa'].unique()
```

```python
df['iq'].unique()
df['iq']=df['iq'].replace(np.nan,df['iq'].median())
df['iq'].unique()
```

```python
df['placement'].unique()
df['placement'].value_counts()
```

```python
sns.scatterplot(x=df['cgpa'],y=df['iq'], hue=df['placement'])
```

```python
## Input and Output variables
x=df.iloc[:,0:2]
y=df.iloc[:,-1]
```
