```python
def step(z):
    return 1 if z>0 else 0
```

```python
def perceptron(x,y):
    X=np.insert(x,0,1,axis=1) ## Bias add kia humne
    weights=np.ones(X.shape[1]) ## Weights add kia humne 
    lr=0.1 ## learning rate liya ha humne

    for i in range(1000):
        j= np.random.randint(0,100)
        y_hat= step(np.dot(x[j], weights))
        weights = weights + lr*(y[j]-y_hat)*x[j]
    return weights[0], weights[1:]
```
