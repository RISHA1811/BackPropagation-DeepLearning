```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
scaler.fit_transform(x)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Dense
```

```python
## Abhi object ko initialize karna ha 
model=Sequential()
```

```python
model.add(Dense(2,activation='sigmoid', input_dim=2))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',metrics=['accuracy']) ## dekh mein yaha batch_size de raha esliye ye ek batch gradient descent hoga , agar mein batch_size=1 daalta toh vo ek sochisticated gradient descent hoga
history = model.fit(x_train,y_train,epochs=50,batch_size=300)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['loss'])
```
